# Pivot y Melt

```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv')
df.head()
```

Ahora, supongamos que queramos visualizar los libros publicados según su género a lo largo de los años. Para esto se usaría un pivote.

```python
pivot = df.pivot_table(index='Genre',columns='Year', values='User Rating',aggfunc='count')
pivot
```

Si quisiéramos ver los libros lanzados en 2009, podríamos hacerlo con `loc`.

```python
pivot.loc[:, 2009]
```

O, por ejemplo, para ver la cantidad de libros que sacó cada autor a lo largo de los años, podemos usar esta tabla pivote.

```python
df.pivot_table(index='Author', columns='Year', values='Name', aggfunc='count').fillna('-')
```

De no hacer el `fillna`, se visualizarían muchos nulos. También podemos llenarlos con 0.

```python
df.melt(id_vars=['Author', 'Year'], value_vars='Name')
```
