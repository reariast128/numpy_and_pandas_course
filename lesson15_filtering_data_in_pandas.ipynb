# Filtrando datos en Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv')
df.head()
```

Podemos empezar como lo hacíamos en NumPy, trabajando con máscaras.

```python
df[df['Year'] > 2016]
```

También, podemos concatenarlas como siempre. Además, para negar el resultado podemos usar la virgulilla ``~``.

```python
df[~(df['Genre'] == 'Fiction') & (df['Year'] > 2016)]
```
