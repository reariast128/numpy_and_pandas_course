A la hora de combinar DataFrames, se sigue una lógica muy similar a la de SQL.



Hay una cosa a tener en cuenta, y es que si existen columnas compartidas, hay que escoger una a la que darle prioridad. Esto es, el tipo de join. Existen 4 tipos de join:

## Tipos de joins

### Left join

Este traerá todos los elementos del dataset de la **izquierda**, y la intersección (o elementos en común) del dataset de la izquierda y de la derecha.



### Right join

Este traerá todos los elementos del dataset de la **derecha**, y la intersección (o elementos en común) del dataset de la izquierda y de la derecha.



### Inner join

Este traerá únicamente la intersección del dataset de la izquierda y de la derecha.



### Outer join

Esto se trae todo, tanto los datos de la izquierda como los de la derecha.



### Consideraciones

En todos estos casos, la unión (o merge) haría que el dataset crezca horizontalmente.

## Concats

### Por filas (axis 0)

En este tipo de merge, el dataset crece verticalmente. Se usa cuando la información de las columnas es igual entre ambos datasets.



### Por columnas (axis 1)

En este tipo de merge, el dataset crece diagonalmente. Se usa cuando ambas bases de datos describen algo de un dato, pero cada una describe algo diferente.



En el caso en el que hayan datos faltantes se generarán nulos.
