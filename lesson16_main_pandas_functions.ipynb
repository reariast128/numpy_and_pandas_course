# Principales funciones de Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv')
df.head()
```

## ``DataFrame.info()``

Esto devuelve una tabla con la información de los tipos de datos que se manejan en el DataFrame, la cantidad de registros, las columnas, entre o el uso de memoria, entre otros.

```python
df.info()
```

## ``DataFrame.describe()``

``describe()`` da información estadística sobre las columnas numéricas del DataFrame.

```python
df.describe()
```

## ``DataFrame.memory_usage()``

Este método nos da información en bytes sobre el uso de memoria del DataFrame.

```python
df.memory_usage()
```

## ``DataFrame[column].value_counts()``

Este método nos indica la cantidad de veces que se repite un valor en una columna.

```python
df['Author'].value_counts()
```

## DataFrame.drop_duplicates()

Este método remueve los duplicados, tanto a nivel de columnas como de filas.

```python
df = pd.concat((df, df))
df
```

```python
df.drop_duplicates(inplace=True)
df
```

## DataFrame.sort_values(column)

Sirve para ordenar los registros de un DataFrame en base a una columna.

```python
df.sort_values('Author', inplace=True)
df
```
